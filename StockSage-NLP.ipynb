```python
import pandas as pd
import sklearn
```

```python
dataset = pd.read_csv('Data.csv' , encoding= 'ISO-8859-1')
#This dataset was on kaggle scrapped from yahoo finance#
```

```python
dataset.head()
#so this dataset has some date value and corresponding top 25 news topics, label values 1 & 0s (1 for stock price increased, 0 means it stayed unchanged or decreased) 
```

```python
dates = dataset['Date'].tolist()
print(dates)
```

```python
#so after kowning this that its contains data within 4101 B-days we will devide the dataset into aprox 80%-20% for one time then 90%-10% traning and testing
#80-20 [2000/01/01 - 2013/03/31 to 2013/04/01 - 2016/07/01]
#90-10 [2000/01/01 - 2014/12/31 to 2015/01/01 - 2016/07/01]
train = dataset[dataset['Date'] < '20150101']
test = dataset[dataset['Date'] > '20141231']
```

```python
#remove puncuations
data = train.iloc[:,2:27]
data.replace("[^a-z A-Z]"," ", regex=True, inplace=True)
```

```python
#converting headlines to lower case to work with bg of words
for i in data :
    data[i] = data[i].str.lower()
```

```python
#joinning the headlines to process
# Step 1: Create an empty list to store each combined headline
headlines = []

# Step 2: Loop through each row in the dataset
for row in range(len(data)):
    # Step 3: Select columns 0 to 24 for the current row
    row_data = data.iloc[row, 0:25]
    
    # Step 4: Convert each value in row_data to a string
    string_values = [str(x) for x in row_data]

    # Step 5: Join all the string values with a space in between
    headline = ' '.join(string_values)
    # Step 6: Append the combined headline to the headlines list
    headlines.append(headline)
```

```python
headlines[0]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.ensemble import RandomForestClassifier
```

```python
## implement BAG OF WORDS
countvector=CountVectorizer(ngram_range=(2,2))
traindataset=countvector.fit_transform(headlines)
```

```python
# implement RandomForest Classifier
randomclassifier=RandomForestClassifier(n_estimators=200,criterion='entropy')
randomclassifier.fit(traindataset,train['Label'])
```

```python
## Predict for the Test Dataset
test_transform= []
for row in range(0,len(test.index)):
    test_transform.append(' '.join(str(x) for x in test.iloc[row,2:27]))
test_dataset = countvector.transform(test_transform)
predictions = randomclassifier.predict(test_dataset)
```

```python
## Import library to check accuracy
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score
matrix=confusion_matrix(test['Label'],predictions)
print(matrix)
score=accuracy_score(test['Label'],predictions)
print(score)
report=classification_report(test['Label'],predictions)
print(report)
```
